```python
import pandas as pd
import numpy as np

# --- 1. Volúmenes Generales ---
N_TOTAL = 100000
PCT_FRAUDE = 0.2  # 10%

N_FRAUDES = int(N_TOTAL * PCT_FRAUDE) 
N_NO_FRAUDES = N_TOTAL - N_FRAUDES        

# --- 2. Rangos Parametrizados para Fraudes (es_fraude = 1) ---

# F1: Fraude manual (Ataque rápido, 2 sucursales, 0 éxitos)
RANGOS_F1 = {'horas': (1, 2), 'intentos': (6, 11), 'sucursales': (2, 5), 'exitosas': (0, 0)}

# F2: Fraude de fuerza bruta (Ataque fulminante, botnet/muchas sucursales, 0 éxitos)
RANGOS_F2 = {'horas': (1, 2), 'intentos': (12, 30), 'sucursales': (3, 10), 'exitosas': (0, 0)}

# --- 3. Rangos Parametrizados para NO Fraudes (es_fraude = 0) ---

# NF1: Comportamiento genuino (A partir de 4 horas, 1 sucursal, éxitos normales)
RANGOS_NF1 = {'horas': (3, 48), 'intentos': (1, 5), 'sucursales': (1, 2), 'exitosas': (1, 5)}

# NF2: Usuario torpe pero genuino (Muchos intentos, PERO alta tasa de éxito. A partir de 4 horas)
RANGOS_NF2 = {'horas': (3, 48), 'intentos': (1, 14), 'sucursales': (1, 1), 'exitosas': (1, 14)}
```

```python
#Funcion para generar datos ficticios

import pandas as pd
import numpy as np

def generar_bloque(n_registros, config, es_fraude):
    np.random.seed(42) # Mantener la reproducibilidad
    df_temp = pd.DataFrame()
    
    # Generamos los datos basados dinámicamente en las tuplas del diccionario
    df_temp['rango_horas'] = np.random.randint(config['horas'][0], config['horas'][1] + 1, n_registros)
    df_temp['cantidad_intentos_transaccion'] = np.random.randint(config['intentos'][0], config['intentos'][1] + 1, n_registros)
    df_temp['cantidad_sucursales_distintas'] = np.random.randint(config['sucursales'][0], config['sucursales'][1] + 1, n_registros)
    
    # Generamos exitosas
    df_temp['cantidad_transacciones_exitosas'] = np.random.randint(config['exitosas'][0], config['exitosas'][1] + 1, n_registros)
    
    # Regla de integridad: Las exitosas nunca pueden ser más que los intentos totales
    df_temp['cantidad_transacciones_exitosas'] = df_temp[['cantidad_transacciones_exitosas', 'cantidad_intentos_transaccion']].min(axis=1)
    
    # Añadimos la etiqueta
    df_temp['es_fraude'] = es_fraude
    
    # Calculamos la tasa de errores consistente
    errores = df_temp['cantidad_intentos_transaccion'] - df_temp['cantidad_transacciones_exitosas']
    df_temp['tasa_errores'] = errores / df_temp['cantidad_intentos_transaccion']
    
    return df_temp
```

```python
# --- 5. Ensamblaje del Dataset ---
bloques = [
    # Inyección de Fraudes (50% F1, 50% F2 asegurando que no se pierdan impares)
    #generar_bloque(N_FRAUDES // 2, RANGOS_F1, 1),
    #generar_bloque(N_FRAUDES // 2 + (N_NO_FRAUDES % 2), RANGOS_F2, 1),

    generar_bloque(1500, RANGOS_F1, 1),
    generar_bloque(3500, RANGOS_F2, 1),
    
    # Inyección de Comportamiento Normal (El 100% de N_NO_FRAUDES usa RANGOS_NF1)
    generar_bloque(N_NO_FRAUDES // 2, RANGOS_NF1, 0),
    generar_bloque(N_NO_FRAUDES // 2 + (N_NO_FRAUDES % 2), RANGOS_NF2, 0),
]

# Unificamos todo en un solo DataFrame
df_sintetico = pd.concat(bloques, ignore_index=True)

# Validación final
print("--- Dimensión y Distribución ---")
print(df_sintetico.shape)
print(df_sintetico['es_fraude'].value_counts())
```

```python
# --- 6. Cálculos Finales ---
# Cálculo de la Tasa de Errores basado en la nueva columna 'cantidad_intentos_transaccion'
errores = df_sintetico['cantidad_intentos_transaccion'] - df_sintetico['cantidad_transacciones_exitosas']
df_sintetico['tasa_errores'] = errores / df_sintetico['cantidad_intentos_transaccion']

# Mezclar (Shuffle) y resetear índice
df_sintetico = df_sintetico.sample(frac=1, random_state=42).reset_index(drop=True)

# Asignar ID único
#df_sintetico.insert(0, 'id', [f"TRX_{i:05d}" for i in range(1, N_TOTAL + 1)])

# Mostrar resultados
df_sintetico.head(30)
```

```python
#Guardar dataset generado en CSV

# 1. Definimos el nombre del archivo de salida
nombre_archivo_csv = 'datasets/dataset_fraude_sintetico.csv'

# 2. Exportamos a CSV
# index=False es CRÍTICO: evita que pandas guarde los números de fila como una columna extra (Unnamed: 0)
# encoding='utf-8' asegura compatibilidad universal, especialmente si tuvieras caracteres especiales
df_sintetico.to_csv(nombre_archivo_csv, index=False, encoding='utf-8')

print(f"✅ ¡Dataset guardado exitosamente como: {nombre_archivo_csv}!")
```

```python

#Division de los dato de entrenamiento

from sklearn.model_selection import train_test_split

# 1. Separar las características predictivas (X) de la variable objetivo (y)
# Eliminamos 'id' porque un identificador no aporta valor predictivo (es ruido)
# Eliminamos 'es_fraude' porque es lo que el modelo debe adivinar
X = df_sintetico.drop(columns=['es_fraude'])
#X = df_sintetico
y = df_sintetico['es_fraude']

# 2. División 80% Entrenamiento (Train) / 20% Prueba (Test)
# stratify=y garantiza que la proporción de fraudes se mantenga igual en ambos lados
# random_state=42 asegura que si vuelves a correr la celda, la división sea exactamente la misma
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.20, 
    random_state=42, 
    stratify=y
)

# 3. Validación Profesional de la División
print("--- Dimensiones de los Conjuntos ---")
print(f"X_train (Entrenamiento): {X_train.shape[0]} filas")
print(f"X_test (Prueba): {X_test.shape[0]} filas")

print("\n--- Proporción de Fraudes en Entrenamiento (y_train) ---")
print(y_train.value_counts(normalize=True) * 100)

print("\n--- Proporción de Fraudes en Prueba (y_test) ---")
print(y_test.value_counts(normalize=True) * 100)

X.head(10)
```

```python
# Guardar dataset en formato parquet

# Exportar a formato Parquet (Estándar de la industria)
nombre_archivo_parquet = 'datasets/dataset_fraude_sintetico.parquet'
df_sintetico.to_parquet(nombre_archivo_parquet, index=False)

print(f"✅ ¡Dataset optimizado guardado como: {nombre_archivo_parquet}!")
```

```python
#Benchmark de algoritmos para el entrenamiento

import pandas as pd
import numpy as np
import time
import warnings
warnings.filterwarnings('ignore') # Para mantener la salida limpia

# Importar los algoritmos
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Importar las métricas
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# 1. Calcular el ratio de desbalanceo para XGBoost
# ratio = cantidad de (No Fraudes) / cantidad de (Fraudes)
ratio_desbalanceo = (y_train == 0).sum() / (y_train == 1).sum()

# 2. Instanciar los modelos con configuraciones para datos desbalanceados
modelos = {
    "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced', n_jobs=-1),
    "XGBoost": XGBClassifier(random_state=42, scale_pos_weight=ratio_desbalanceo, eval_metric='logloss'),
    "LightGBM": LGBMClassifier(random_state=42, class_weight={0: 1, 1: 10}, verbose=-1),
    # Isolation Forest espera la tasa de contaminación (porcentaje de fraudes)
    "Isolation Forest": IsolationForest(random_state=42, contamination=PCT_FRAUDE, n_jobs=-1)
}

# 3. Diccionario para guardar los resultados
resultados = []

print("🚀 Iniciando el Benchmark de Modelos...\n")

# 4. Bucle de Entrenamiento y Evaluación
for nombre, modelo in modelos.items():
    inicio_tiempo = time.time()
    
    # Entrenar el modelo
    if nombre == "Isolation Forest":
        # Isolation Forest no usa y_train para entrenar
        modelo.fit(X_train)
        # Predicción: devuelve -1 para anomalía (fraude) y 1 para normal
        preds_crudas = modelo.predict(X_test)
        # Mapeamos: -1 -> 1 (Fraude), 1 -> 0 (No Fraude)
        y_pred = np.where(preds_crudas == -1, 1, 0)
    elif nombre == "LightGBM":
        reglas_negocio = {
            'rango_horas': -1,                     # Menos tiempo = Más probabilidad de ataque rápido
            'cantidad_intentos_transaccion': 1,    # Más intentos = Más probabilidad de fuerza bruta
            'cantidad_transacciones_exitosas': -1, # Más éxitos = Menos riesgo (comportamiento genuino)
            'cantidad_sucursales_distintas': 1,    # Más sucursales = Más probabilidad de ataque distribuido
            'tasa_errores': 1                      # Más errores = Más riesgo
        }
        monotone_tuple = tuple(reglas_negocio.get(col, 0) for col in X_train.columns)

        modelo.set_params(monotone_constraints=monotone_tuple)
        
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_test)
    else:
        # Modelos supervisados normales
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_test)
        
    tiempo_entrenamiento = time.time() - inicio_tiempo
    
    # 5. Calcular Métricas
    roc_auc = roc_auc_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    # Guardar en la lista
    resultados.append({
        "Modelo": nombre,
        "ROC-AUC": roc_auc,
        "Recall (Atrapa Fraudes)": recall,
        "Precision (Exactitud Alerta)": precision,
        "F1-Score": f1,
        "Tiempo (segundos)": tiempo_entrenamiento
    })
    
    print(f"✅ {nombre} evaluado en {tiempo_entrenamiento:.2f}s")

# 6. Crear un DataFrame con los resultados y ordenarlo por F1-Score (la métrica más balanceada)
df_benchmark = pd.DataFrame(resultados).sort_values(by="F1-Score", ascending=False).reset_index(drop=True)

# 7. Mostrar resultados estilizados
print("\n🏆 Resultados Finales del Benchmark:")
display(df_benchmark.style.background_gradient(cmap='viridis', subset=["ROC-AUC", "Recall (Atrapa Fraudes)", "Precision (Exactitud Alerta)", "F1-Score"]))
```

```python
#Aunque el algoritmo LightGBM tiene mejor respuesta, no arroja datos razonables

#Heatmap de Isolation forest

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier, IsolationForest

# 1. Asignamos el modelo Isolation Forest
# (Asegúrate de usar el nombre correcto de tu variable, ej. iso_model o modelos["Isolation Forest"])
modelo = IsolationForest(random_state=42, contamination=PCT_FRAUDE, n_jobs=-1)

modelo.fit(X_train)
# 2. Generamos las predicciones "crudas" del modelo
preds_crudas = modelo.predict(X_test)

# 3. 🌟 PASO CRÍTICO: Mapeo de etiquetas
# Convertimos: -1 (Anomalía) -> 1 (Fraude) | 1 (Normal) -> 0 (Genuino)
y_pred_iso = np.where(preds_crudas == -1, 1, 0)

# 4. Calculamos la matriz de confusión con las etiquetas ya corregidas
cm = confusion_matrix(y_test, y_pred_iso)

# 5. Configuramos y renderizamos el Heatmap
plt.figure(figsize=(8, 6))

# Cambié cmap='Blues' por 'Oranges' para diferenciar los reportes de los modelos
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', cbar=False,
            xticklabels=['Genuino (0)', 'Fraude (1)'],
            yticklabels=['Genuino (0)', 'Fraude (1)'])

plt.xlabel('Predicción del Modelo (Isolation Forest)', fontsize=12, fontweight='bold')
plt.ylabel('Dato Real', fontsize=12, fontweight='bold')
plt.title('Matriz de Confusión - Detección de Fraude (Isolation Forest)', fontsize=14)
plt.show()
```

```python
#Crear el artefacto del modelo LightGBM entrenado

import joblib

# 1. Definir el diccionario de Reglas de Negocio
# 1  = Relación Positiva (Si la variable sube, el riesgo de fraude SUBE)
# -1 = Relación Negativa (Si la variable sube, el riesgo de fraude BAJA)
# 0  = Sin restricción (El modelo decide libremente)
reglas_negocio = {
    'rango_horas': -1,                     # Menos tiempo = Más probabilidad de ataque rápido
    'cantidad_intentos_transaccion': 1,    # Más intentos = Más probabilidad de fuerza bruta
    'cantidad_transacciones_exitosas': -1, # Más éxitos = Menos riesgo (comportamiento genuino)
    'cantidad_sucursales_distintas': 1,    # Más sucursales = Más probabilidad de ataque distribuido
    'tasa_errores': 1                      # Más errores = Más riesgo
}

# 2. Alinear las reglas con el orden exacto de las columnas de tu dataset 'X'
# LightGBM requiere que las restricciones sean una tupla ordenada, ej: (1, -1, 1, 1, -1)
# Usamos un list comprehension para mapear el diccionario automáticamente
monotone_tuple = tuple(reglas_negocio.get(col, 0) for col in X.columns)

print(f"Restricciones aplicadas en el orden de X: {monotone_tuple}\n")

display(X.info())

# 3. Extraer el modelo y actualizar sus parámetros
modelo_ganador = modelos["LightGBM"]
modelo_ganador.set_params(monotone_constraints=monotone_tuple)

# 4. Re-entrenar con el 100% de los datos y la nueva regla irrompible
print("Re-entrenando LightGBM con restricciones monotónicas...")
modelo_ganador.fit(X, y)

# 5. Guardar el modelo blindado en el disco duro
nombre_artefacto = 'modelo_lightgbm_fraudes.joblib'
joblib.dump(modelo_ganador, nombre_artefacto)

print(f"✅ ¡Artefacto generado con éxito: {nombre_artefacto}!")
print("Este archivo garantiza que la IA respete tu lógica de negocio en producción.")

X.head(1)
```
