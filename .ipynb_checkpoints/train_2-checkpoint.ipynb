```python
import pandas as pd
import joblib
from sklearn.ensemble import IsolationForest

# 1. Cargar el dataset desde la ruta especificada
df = pd.read_csv('datasets/dataset_fraude_sintetico.csv')

# 2. Selección de variables para el entrenamiento
# Eliminamos 'id' (no aporta valor) y 'es_fraude' (es supervisado, y este es un modelo no supervisado)
features = ['rango_horas', 'cantidad_intentos_transaccion', 'cantidad_transacciones_exitosas', 
            'cantidad_sucursales_distintas', 'tasa_errores']

X = df[features]

print(f"Dataset cargado con éxito. Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
print("Variables utilizadas para el entrenamiento:", features)
```

```python
# 1. Instanciar el modelo
# contamination=0.012 es el porcentaje estimado de fraudes (1.2%)
iso_model = IsolationForest(
    n_estimators=100, 
    contamination=0.1, 
    random_state=42,
    n_jobs=-1
)

# 2. Entrenar el modelo
print("Entrenando el modelo Isolation Forest...")
iso_model.fit(X)
print("¡Entrenamiento finalizado!")

# 3. Verificación rápida (Inferencia)
# -1 indica anomalía (posible fraude), 1 indica dato normal
predictions = iso_model.predict(X)
n_anomalias = list(predictions).count(-1)

print(f"Se han detectado {n_anomalias} posibles anomalías en el dataset.")
```

```python
# Guardar el artefacto en formato .joblib
nombre_archivo = 'modelo_iso_forest_2.joblib'
joblib.dump(iso_model, nombre_archivo)

print(f"✅ Modelo guardado exitosamente como '{nombre_archivo}'.")
```
