```python
import pandas as pd
import numpy as np

# 1. Cargar el dataset especificando la columna de fecha para su parseo automático
# Nota: Si el formato de fecha es especial, pandas suele detectarlo, de lo contrario podemos especificarlo.
try:
    df = pd.read_csv('datasets/cotejos.csv', parse_dates=['log_fecha'])
    print("¡Dataset cargado exitosamente!")
except KeyError:
    # En caso de que la columna tenga otro nombre exacto en el CSV
    df = pd.read_csv('cotejos.csv')
    print("Dataset cargado. Nota: Verifica el nombre exacto de la columna de fecha para parsearla.")

# 2. Inspección inicial de la estructura y tipos de datos
print("\n--- Dimensiones del Dataset ---")
print(f"Registros totales: {df.shape[0]} | Columnas: {df.shape[1]}")

print("\n--- Tipos de Datos e Incompletitud (Valores Nulos) ---")
display(df.info())

print("\n--- Primeros 5 registros ---")
display(df.head())
```

```python
# Ordenar cronológicamente por documento y tiempo, esto es vital para análisis de ventanas
df = df.sort_values(by=['log_id', 'log_fecha']).reset_index(drop=True)

df.head(10)
```

```python
#Eliminar columnas con todos los valores nulos
df_clean = df.dropna(axis=1, how='all')
print(f"Columnas eliminadas. Nueva forma: {df_clean.shape}")
```

```python
df_clean.head(10)
```

```python
import pandas as pd
import numpy as np

df = df_clean

# Asumiendo que tu dataframe se llama 'df' (o 'df_clean' si ejecutaste el paso anterior)

# 1. Calculamos el tamaño de cada grupo definido por 'log_data'
# transform('size') devuelve una serie del mismo tamaño que el dataframe original 
# con la cantidad de elementos que tiene el grupo al que pertenece cada fila.
tamano_grupo = df.groupby('log_data')['log_data'].transform('size')

# 2 y 3. Lógica para asignar las columnas 'request' y 'response'
# Dado que en ambos casos (1 fila o 2 filas) 'request' siempre es 1, lo asignamos directamente:
df['request'] = 1

# Para 'response', será 1 si el tamaño del grupo es 2, y 0 si el tamaño del grupo es 1.
# Convertimos el resultado booleano (True/False) a entero (1/0) usando astype(int)
df['response'] = (tamano_grupo == 2).astype(int)


# 4. (Opcional pero recomendado) Validar que la operación fue exitosa
print("--- Muestra de las nuevas características creadas ---")
df.head(10)
```

```python
# Aplicamos np.nan (valor nulo estándar en computación científica) 
# a la columna 'log_mens_respuesta' exclusivamente para las filas donde 'log_tipo' es 'REQ'
df.loc[df['log_tipo'] == 'REQ', 'log_mens_respuesta'] = np.nan

df.head(10)
```

```python
# 1. Definimos las columnas que queremos unificar de la fila 2 (RES) a la fila 1 (REQ)
columnas_a_unificar = ['log_sucursal', 'log_hora', 'log_nut', 'log_cod_respuesta', 'log_mens_respuesta']

# 2. Agrupamos por 'log_data' y aplicamos un llenado bidireccional (bfill y ffill).
# bfill() copia los datos de abajo hacia arriba (de RES a REQ).
# ffill() los copia de arriba hacia abajo (por si alguna vez RES está antes que REQ).
df[columnas_a_unificar] = df.groupby('log_data')[columnas_a_unificar].bfill().ffill()

df.head(10)
```

```python
#Eliminar filas de tipo RES

df = df[df['log_tipo'] != 'RES'].copy()

# 2. Reiniciar el índice
# Al eliminar filas, los índices originales se conservan (ej: 0, 1, 3, 5...). 
# Reiniciarlos asegura que vuelvan a ser secuenciales, evitando errores en futuras iteraciones.
df = df.reset_index(drop=True)

df.head(40)
```
